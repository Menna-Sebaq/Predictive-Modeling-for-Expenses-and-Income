```python
data_encoded = pd.get_dummies(data_filtered, dtype=int)
data_encoded
```

```python
X = data_encoded.drop(data_encoded.columns[2], axis=1).values
y = data_encoded.iloc[:, 2:3].values

print('x',X)

print('y',y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.2, random_state = 1)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(X_train, y_train)
y_predict = lg.predict(X_test)
```

```python
plt.scatter(X_test[:, 0], y_test, color='red', label='Actual') 
plt.scatter(X_test[:, 0], y_predict, color='blue', label='Predicted')
plt.title('Predicted vs Actual')
plt.xlabel('expenses')
plt.ylabel('age')
plt.legend()
plt.show()
```

```python
plt.scatter(X_test[:, 1], y_test, color='red', label='Actual')  # Considering the first column of X_test
plt.scatter(X_test[:, 1], y_predict, color='blue', label='Predicted')
plt.title('Predicted vs Actual')
plt.xlabel('expenses')
plt.ylabel('bmi')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_predict)
mse = mean_squared_error(y_test, y_predict)
r_squared = r2_score(y_test, y_predict)

print(f"Mean Absolute Error: {mae:.4f}\n")
print(f"Mean Squared Error: {mse:.4f}\n")
print(f"R-squared: {r_squared:.4f}\n")
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
import numpy as np

# Reshape y_train to a 1D array 
y_train_reshaped = np.ravel(y_train)

gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
cv_scores = cross_val_score(gb_model, X_train, y_train_reshaped, cv=5, scoring='r2')

print("Cross-validation R-squared scores:", cv_scores)
print(f"Mean R-squared: {cv_scores.mean():.4f}")
```
